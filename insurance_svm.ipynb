```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder=OrdinalEncoder(categories=[['female','male']])
dataset[['sex']]=ordinal_encoder.fit_transform(dataset[['sex']])
```

```python
dataset[['sex']]=dataset[['sex']].astype(int)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder=OrdinalEncoder(categories=[['yes','no']])
dataset[['smoker']]=ordinal_encoder.fit_transform(dataset[['smoker']])
```

```python
dataset[['smoker']]=dataset[['smoker']].astype(int)
```

```python
dataset.head()
```

```python
independent=dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel='poly',C=1)
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

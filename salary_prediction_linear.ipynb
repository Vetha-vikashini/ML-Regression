```python
import pandas as pd #pandas are used handle the dataset
```

```python
dataset=pd.read_csv('Salary_Data.csv') #by using read_csv method,here we  uploadin the data set,ie., it is data collection process 
```

```python
dataset.head() #displaying first 5 rows of dataset
```

```python
independent=dataset[['YearsExperience']] #seperate the input from dataset and store it in variable independent
```

```python
dependent=dataset[['Salary']] #seperate the output from dataset and store it in variable dependent
```

```python
from sklearn.model_selection import train_test_split #importing method train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
#by using the method we split the train and test data 
```

```python
from sklearn.linear_model import LinearRegression #import the linear regression model formula in the form of class
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train) #creating the model by using the fit method
```

```python
weight=regressor.coef_        #taking the weight or slope from the equation y=wx+b
print("WEIGHT:",weight)
```

```python
bias=regressor.intercept_    #taking the bias or intercept from the equation y=wx+b
print("INTERCEPT:",bias)
```

```python
y_pred=regressor.predict(x_test)   #predicting the output for data in x_test
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)     #calculating the performance or perfection of model using r2 validating parameter
```
